## Treating Outliers

We will use Capping and Flooring.

So 1st we need to identify and the <b>99th</b> & <b>1</b> percentile values of this 2 variables.

The logic is that, if the value is More that <b>99th</b> or Less than <b>1</b> Percentile then it is an outlier

```python
np.percentile(data.n_hot_rooms,[99])    # Since Outliers is at higher end
```

```python
# To fetch 1st value in an array

np.percentile(data.n_hot_rooms,[99])[0]
```

```python
uv = np.percentile(data.n_hot_rooms,[99])[0]
```

```python
data[(data.n_hot_rooms)>uv]       # This will remove outliers
```

```python
data.n_hot_rooms[(data.n_hot_rooms)>3*uv]
```

```python
data.n_hot_rooms[(data.n_hot_rooms)>3*uv] = 3*uv
```

```python
data[(data.n_hot_rooms)>uv]
```

```python
np.percentile(data.rainfall,[1])[0]     # Since Outlier is at lower end
```

```python
lv = np.percentile(data.rainfall,[1])[0]
```

```python
data[(data.rainfall)<lv]
```

```python
data.rainfall[(data.rainfall)<0.3*lv] = 3*lv
```
